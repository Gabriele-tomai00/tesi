```python
import json
import os

# Percorso del file JSON
filename = 'orari_lezioni/resp.json'

# Verifica che il file esista
if not os.path.exists(filename):
    print(f"Il file '{filename}' non Ã¨ stato trovato.")
else:
    # Carica il contenuto del file JSON
    with open(filename, 'r', encoding='utf-8') as file:
        data = json.load(file)

    # Estrai curricula_course e file_timestamp
    curricula_course = data.get('curricula_course')
    file_timestamp = str(data.get('file_timestamp'))  # Convertito in stringa

    print(f"curricula_course: {curricula_course}")
    print(f"file_timestamp: {file_timestamp}")
    print("\n")
    print(f"  anno_accademico: {data.get('anno')}")
    print(f"  settimana_del: {data.get('first_day')}")

    # Cerca l'elemento con timestamp corrispondente
    matched = None
    for entry in data.get('date_inizio_fine_curriculum', []):
        timestamp_value = entry.get('timestamp', {}).get('0')
        if str(timestamp_value) == file_timestamp:
            matched = entry
            break

    # Stampa l'elemento corrispondente
    if matched:
        date_info = matched.get('date', {}).get('1', {})
        print("\nElemento corrispondente:")
        print(f"  AnnoCorso: {date_info.get('AnnoCorso')}")
        print(f"  DataInizio: {date_info.get('DataInizio')}")
        print(f"  DataFine: {date_info.get('DataFine')}")
        
    else:
        print("\nNessun elemento con timestamp corrispondente trovato.")
```

FILTRO CELLE CON LE INFO DELLE LEZIONI

```python
import json

# Lista delle chiavi da mantenere
CHIAVI_DA_TENERE = [
    "codice_insegnamento",
    "nome_insegnamento",
    "data",
    "codice_aula",
    "codice_sede",
    "aula",
    "orario",
    "tipo",
    "ora_inizio",
    "ora_fine",
    "Annullato",
    "codice_docente",
    "docente",
]

filename = "response_grid/resp.json"

# Carica il JSON
with open(filename, "r", encoding="utf-8") as f:
    data = json.load(f)

# Filtra le celle
celle_filtrate = [
    {k: cella[k] for k in CHIAVI_DA_TENERE if k in cella}
    for cella in data.get("celle", [])
]

# Risultato finale
data_filtrato = {"orario_lezioni": celle_filtrate}

# Salva il JSON filtrato
with open("response_grid/resp_filtrato.json", "w", encoding="utf-8") as f:
    json.dump(data_filtrato, f, indent=4, ensure_ascii=False)

# Stampa a video
print(json.dumps(data_filtrato, indent=4, ensure_ascii=False))
```
