IMPOSTO IL webdriver

```python
import json
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

# Imposta Chrome headless
chrome_options = Options()
chrome_options.add_argument("--headless=new")
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--window-size=1920,1080")

driver = webdriver.Chrome(options=chrome_options)

URL_FORM = "https://orari.units.it/agendaweb/index.php?view=easycourse&_lang=it&include=corso"
driver.get(URL_FORM)
```

Estraggo i valori della tendina 'Anno accademico' e imposto quello più recente

```python
# Trova la tendina anno
select_anno_scolastico = driver.find_element(By.ID, "cdl_aa")

# Estrai coppie (value, label) ignorando la prima opzione vuota
values_anni = [
    {"value": opt.get_attribute("value"), "label": opt.text}
    for opt in select_anno_scolastico.find_elements(By.TAG_NAME, "option")
    if opt.get_attribute("value")
]

print(json.dumps(values_anni, indent=2, ensure_ascii=False))


from selenium.webdriver.support.ui import Select

# Trova l'anno più recente (valore massimo)
latest_value = max(v["value"] for v in values_anni)
print(f"\nAnno più recente trovato: {latest_value}")

# Seleziona quell'anno nel menu
select = Select(select_anno_scolastico)
select.select_by_value(latest_value)

print(f"Impostato l'anno {latest_value} sul form!")
time.sleep(0.2)

anno_scolastico = latest_value
```

Funzione per creare l'url

```python
from urllib.parse import urlencode, quote

def build_orario_url(anno_scolastico, scuola, corso, anno2, date, lang="it"):
    base_url = "https://orari.units.it/agendaweb/index.php"

    params = {
        "view": "easycourse",
        "form-type": "corso",
        "include": "corso",
        "txtcurr": "",
        "anno": anno_scolastico,
        "scuola": scuola,
        "corso": corso,
        "anno2[]": anno2,  # verrà encodato
        "visualizzazione_orario": "cal",
        "date": date,
        "periodo_didattico": "",
        "_lang": lang,
        "list": "",
        "week_grid_type": "-1",
        "ar_codes_": "",
        "ar_select_": "",
        "col_cells": "0",
        "empty_box": "0",
        "only_grid": "0",
        "highlighted_date": "0",
        "all_events": "0",
        "faculty_group": "0",
    }

    return f"{base_url}?{urlencode(params, quote_via=quote)}"

url = build_orario_url(anno_scolastico, scuola, corso, anno2, date="2024-09-23")
print(url)
```
